```python

df = pd.read_csv("Housing.csv")   

# Binary Encoding
binary_cols = [
    'mainroad',
    'guestroom',
    'basement',
    'hotwaterheating',
    'airconditioning',
    'prefarea'
]

for col in binary_cols:
    read_data[col] = df[col].map({'yes': 1, 'no': 0})

# One Hot Encoding
read_data = pd.get_dummies(read_data, columns=['furnishingstatus'], drop_first=True)

print(read_data.head())
print(read_data.info())
```

```python
import pandas as pd

data = {
    'Name': ['Ali', 'Sara', 'Omar', 'Mona', 'Youssef'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
    'City': ['Cairo', 'Giza', 'Cairo', 'Alex', 'Giza'],
    'Level': ['High', 'Medium', 'Low', 'Medium', 'High'],
    'Salary': [5000, 7000, 6000, 6500, 7200]
}

df = pd.DataFrame(data)

print(df)
```

```python

from sklearn.preprocessing import LabelEncoder

df_lable = df.copy()

le = LabelEncoder()

for col in ['Gender', 'City', 'Level']:
    df_lable[col] = le.fit_transform(df_lable[col])

    print(df_lable)
```

```python
df_one_hot = pd.get_dummies(
    df, 
    columns =['Gender', 'City', 'Level'],

    drop_first= True

)

print(df_one_hot)
```

```python
df_freq = df.copy()

for col in ['Gender', 'City', 'Level']:
    freq = df_freq[col].value_counts()
    df_freq[col] = df_freq[col].map(freq)

print(df_freq)
```

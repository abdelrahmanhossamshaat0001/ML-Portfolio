```python
prices_input = input("Enter product prices separated by spaces: ")
prices_list = prices_input.split()

for price_str in prices_list:
    price = float(price_str)
    
    if price >= 0 and price <= 50:
        print(f"{price} is Cheap")
        
    elif price > 50 and price <= 100:
        print(f"{price} is Moderate") 
        
    elif price > 100:
        print(f"{price} is Expensive")
        
    else:
        print(f"{price} is Invalid Price")
```

```python
prices = list(map(float, input("Enter the prices separated by spaces: ").split()))

for price in prices:
    if price <= 50:
        print(f"{price} is Cheap")
    elif price > 50 and price <= 100:
        print(f"{price} is mid")
    else:
        print(f"{price} is Expensive")
```

```python
text = input("Enter an even-length string: ")


step2 = text[::2]

mid_index = len(step2) // 2
first_half = step2[:mid_index]
last_half = step2[mid_index:]
reversed_last_half = last_half[::-1] 
step3 = first_half + reversed_last_half


step4 = step3[-1] + step3[1:-1] + step3[0]


final_result = step4.upper()

print("input:", text)
print("output:", final_result)
```

# Exercise 4


```python

n = 3
grid = [[0, 0, 0] for _ in range(n)]

for i in range(n):
    for j in range(n):
        
        if i == j and i + j == n - 1:
            grid[i][j] = 3
            
        elif i == j:
            grid[i][j] = 1
            
        elif i + j == n - 1:
            grid[i][j] = 2

print("Example of output:\n")
for row in grid:
    print(*row)
```

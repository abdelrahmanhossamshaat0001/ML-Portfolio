**Conditions Using where**

**np.where():** It returns indices where a condition is True, or it can be used to choose values based on a condition (like an if-else for arrays).

````python
a = np.array([10, 20, 30, 40, 50])
indices = np.where(a > 25)

print(indices)
print(a[indices])
a[0]
````

**Excercise 2: Sales Data Analysis**
- Create a 4×7 array of random sales between $100 and $500.

- Apply a 10% tax to sales exceeding $300. (x1.10)

- Use np.where() to give a 5% discount to sales below $200. (x 0.95)

- Find the average sales for weekends (last 2 columns).

- Identify days with sales > $400 (boolean mask).




`Solution:`

````python
sales = np.random.uniform(100, 500, (4, 7))
sales_with_tax = np.where(sales > 300, sales * 1.10, sales)
sales_final = np.where(sales_with_tax < 200, sales_with_tax * 0.95, sales_with_tax)
weekend_avg = sales_final[:, -2:].mean()
high_sales_mask = sales_final > 400
````
